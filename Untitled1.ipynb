### (Análisis 2) - ENCODING

================================================

* La tarea para esta sección es:

  1. Identificar variables categóricas en el dataset (tipo object).
  2. Aplicar one-hot enconding (codificación dummy) para transformar cada categoría en una columna numérica.
  3. Mantener la variable objeto (Churn) sin codificar, si ya está en 0/1.

  Debajo va el bloque de código:

```python
# Detectar las columnas categóricas
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Revisar tipos de datos de cada columna categórica
df[categorical_cols].applymap(type).nunique()
```

Lo que primero hice es revisar qué columnas tienen diccionarios. La devolución que me dió confirma que todas las columnas contienen diccionarios.
Para trabajar, primero tengo que normalizar las columnas:

```python
from pandas import json_normalize

# Normalizar cada columna que contiene diccionarios
df_customer = json_normalize(df['customer'])
df_phone = json_normalize(df['phone'])
df_internet = json_normalize(df['internet'])
df_account = json_normalize(df['account'])
```

Ahora, concatenar los resultados y eliminar columnas originales:

```python
df = pd.concat([df.drop(columns=['customer', 'phone', 'internet', 'account']),
                df_customer, df_phone, df_internet, df_account], axis=1)
```

Ahora el dataframe ya no tiene columnas con diccionarios, unicamente trabajamos con valores planos (strings, números, booleanos).

El próximo paso es aplicar el One-Hot Encoding:

```python
# Identificar columnas categóricas que quedan (strings/objetos)
categorical_cols = df.select_dtypes(include=['object']).columns

# Aplicar one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Verificación
print("Dimensiones después de codificación:", df_encoded.shape)
df_encoded.head()
```

### (Análisis 8) - MODELADO PREDICTIVO - SEPARACIÓN DE DATOS:

========================================

Según la tarea, hay que dividir los datos en entrenamiento y prueba, por ejemplo 70/30.
Como la variable objetivo ya está en Churn_bin, podemos usar train_test_split de scikit-learn.

Paso al bloque de código:

```python
from sklearn.model_selection import train_test_split

# Variables predictoras (features)
X = df_encoded.drop(columns=['Churn_bin', 'Churn_No', 'Churn_Yes'])  # quitamos la variable objetivo y las derivadas

# Variable objetivo
y = df_encoded['Churn_bin']

# División en entrenamiento y prueba (70% / 30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Ver tamaños
print("Tamaño X_train:", X_train.shape)
print("Tamaño X_test:", X_test.shape)
print("Tamaño y_train:", y_train.shape)
print("Tamaño y_test:", y_test.shape)
```

* Notas importantes:

1. stratify=y asegura que la proporción de churn se mantenga en ambos conjuntos.
2. Se puede ajustar test_size.
3. Quité las columnas Churn_No y Churn_Yes porque ya representé en la clase con Churn_bin.

### (Análisis 9) - CREACIÓN DE MODELOS

======================================

La tarea en esta sección es crear al menos dos modelos distintos, uno que requiera normalización (Regresión Logística o KNN) y otro que no (Árbol de Decisión o Random Forest).

Con este dato, vamos al código:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# ===============================
# 1. Normalización para modelos sensibles a escala (Regresión Logística)
# ===============================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Modelo 1: Regresión Logística
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train_scaled, y_train)
y_pred_logreg = logreg.predict(X_test_scaled)

print("=== Regresión Logística ===")
print("Accuracy:", accuracy_score(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))

# ===============================
# 2. Modelo que no requiere normalización (Random Forest)
# ===============================
rf = RandomForestClassifier(random_state=42, n_estimators=100)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

Voy a una explicación rápida de lo que hice:

1. Regresión logística:
*  Sensible a la escala de datos.
* Normalicé con StandardScaler para que las variables tengan media 0 y desviación estándar 1.
Esto asegura que los coeficientes no estén sesgados por magnitudes distintas.

2. Random Forest:
* Basado en árboles, no depende de la escala.
* No requiere normalización.

3. Evaluación:
* accuracy_score y classification_report muestra desempeño, precisión, recall y F1-score.

Ahora voy a agregar la Matriz de confusión y un pequeño análisis critico comparando los modelos:

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# ===============================
# Matriz de confusión - Regresión Logística
# ===============================
cm_logreg = confusion_matrix(y_test, y_pred_logreg)
plt.figure(figsize=(5,4))
sns.heatmap(cm_logreg, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión - Regresión Logística')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

# ===============================
# Matriz de confusión - Random Forest
# ===============================
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(5,4))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens')
plt.title('Matriz de Confusión - Random Forest')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()
```

1. Comparar desempeño de los modelos

* Después de entrenar estos modelos y evaluar métricas como accuracy, precisión, recall y F1-score, prestar atención en lo siguiente:

* Exactitud (Accuracy): porcentaje de predicciones correctas sobre el total.

a) Si un modelo tiene muy alta exactitud pero las clases están desbalanceadas, podría ser engañoso.

* Precisión (Precision): de todos los clientes que el modelo predijo como “churn”, cuántos realmente cancelaron.

* Recall: de todos los clientes que realmente cancelaron, cuántos el modelo detectó correctamente.

* F1-score: combinación de precisión y recall, útil cuando hay desbalance de clases.

En este caso, como la clase “churn” suele ser minoritaria, recall y F1-score para churn son métricas muy importantes.

2. Revisar posibles overfitting o underfitting

a) Overfitting:

* El modelo funciona demasiado bien con los datos de entrenamiento pero mal con los datos de prueba.

* Señal de que “memoriza” el conjunto de entrenamiento y no generaliza.

* Causas comunes: modelo muy complejo (p. ej. Random Forest con muchos árboles y profundidad ilimitada).

* Soluciones: limitar la profundidad, aumentar datos de entrenamiento, usar regularización.

b) Underfitting:

* El modelo no logra capturar bien las tendencias ni en entrenamiento ni en test.

* Señal de que es demasiado simple para los datos.

* Causas comunes: modelo lineal para relaciones no lineales, pocos features, poca complejidad.

* Soluciones: aumentar complejidad (más parámetros, más árboles, polinomios, features adicionales).

* Calculé accuracy_score y classification_report, que incluyen exactitud, precisión, recall y F1-score.

### (Análisis 10) - REGRESIÓN LOGÍSTICA: COEFICIENTES

========================================================

Anteriormente habia hecho Regresión Logística y Random Forest, ahora me voy a enfocar en como extraer la relevancia de las variables para cada uno:

```python
# Crear un DataFrame con coeficientes y nombres de variables
coef_logreg = pd.DataFrame({
    'Variable': X_train.columns,
    'Coeficiente': logreg.coef_[0]
})

# Ordenar por importancia (valor absoluto)
coef_logreg['Importancia'] = coef_logreg['Coeficiente'].abs()
coef_logreg.sort_values(by='Importancia', ascending=False, inplace=True)

print("=== Variables más relevantes - Regresión Logística ===")
display(coef_logreg[['Variable', 'Coeficiente']])
```

Los coeficientes (coef_) muestran como cada variable afecta la probabilidad de cancelación.

* Coeficiente positivo indica: aumenta la probabilidad de churn.
* Coeficiente negativo: disminuye la probabilidad de churn.
* Se puede usar un valor absoluto para ordenar la relevancia de mayor a menor.

Random Forest es un algoritmo basado en árboles de decisión, calcula como cada variable reduce la impureza en los nodos del árbol:

```python
# Importancia de variables
importancia_rf = pd.DataFrame({
    'Variable': X_train.columns,
    'Importancia': rf.feature_importances_
})

# Ordenar de mayor a menor
importancia_rf.sort_values(by='Importancia', ascending=False, inplace=True)

print("=== Variables más relevantes - Random Forest ===")
display(importancia_rf)
```

* Interpretación:
1. Valores altos: variable que más contribuyen a la predicción de cancelación.
2. Valores bajos: variables menos relevantes

Voy a escribir un bloque de codigo con un gráfico para que se distinga claramente la relevancia:

```python
# Gráfico barras - Random Forest
plt.figure(figsize=(10,6))
sns.barplot(x='Importancia', y='Variable', data=importancia_rf.head(10))
plt.title('Top 10 variables más relevantes - Random Forest')
plt.show()

# Gráfico barras - Regresión Logística
plt.figure(figsize=(10,6))
sns.barplot(x='Importancia', y='Variable', data=coef_logreg.head(10))
plt.title('Top 10 variables más relevantes - Regresión Logística')
plt.show()
```

### CONCLUSION FINAL:

Armé un bloque de código que va a generar un informe final automático en Markdown, este informe incluye:

* Resumen de los modelos.
* Métricas.
* Variables más importantes.
* Insights y estrategias

Dejo listo para ejecutar el código:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from IPython.display import display, Markdown
import pandas as pd

# ===============================
# Suponemos que tus modelos ya están entrenados:
# logreg -> modelo de regresión logística
# rf -> modelo Random Forest
# X_test, y_test -> datos de prueba
# ===============================

# Calcular métricas para cada modelo
def calcular_metricas(modelo, X, y):
    y_pred = modelo.predict(X)
    return {
        "Accuracy": accuracy_score(y, y_pred),
        "Precision": precision_score(y, y_pred),
        "Recall": recall_score(y, y_pred),
        "F1-score": f1_score(y, y_pred),
        "Confusion Matrix": confusion_matrix(y, y_pred)
    }

metricas_logreg = calcular_metricas(logreg, X_test, y_test)
metricas_rf = calcular_metricas(rf, X_test, y_test)

# Crear DataFrame resumen
df_resumen = pd.DataFrame({
    "Modelo": ["Regresión Logística", "Random Forest"],
    "Accuracy": [metricas_logreg["Accuracy"], metricas_rf["Accuracy"]],
    "Precision": [metricas_logreg["Precision"], metricas_rf["Precision"]],
    "Recall": [metricas_logreg["Recall"], metricas_rf["Recall"]],
    "F1-score": [metricas_logreg["F1-score"], metricas_rf["F1-score"]]
})

# ===============================
# Variables importantes (ajustar según tus resultados)
# ===============================
coef_logreg_top = pd.DataFrame({
    "Variable": ["tenure", "Charges.Total", "Contract"],
    "Coeficiente": [0.8, 0.5, -0.3]  # reemplazar con tus coeficientes reales
})

importancia_rf_top = pd.DataFrame({
    "Variable": ["tenure", "Charges.Monthly", "InternetService"],
    "Importancia": [0.35, 0.25, 0.15]  # reemplazar con tus importancias reales
})

# Función para crear tabla Markdown
def tabla_md(df, columnas):
    tabla = "| " + " | ".join(columnas) + " |\n"
    tabla += "| " + " | ".join(["---"]*len(columnas)) + " |\n"
    for _, row in df.iterrows():
        tabla += "| " + " | ".join([str(row[col]) for col in columnas]) + " |\n"
    return tabla

# ===============================
# Generar informe final
# ===============================
informe_md = f"""
# Informe Final - Desafío Telecom X Parte 2

## 1️⃣ Resumen del análisis
Se trabajó con un dataset de clientes de telecomunicaciones. Se aplicaron técnicas de preprocesamiento, eliminación de columnas irrelevantes, codificación de variables categóricas y análisis de correlación. Se dividió el dataset en conjuntos de entrenamiento y prueba para evaluar los modelos predictivos.

## 2️⃣ Modelos evaluados
- **Regresión Logística:** Normalización aplicada
- **Random Forest:** Sin normalización

## 3️⃣ Resultados de evaluación
El modelo con mejor F1-score fue **{'Random Forest' if metricas_rf['F1-score']>metricas_logreg['F1-score'] else 'Regresión Logística'}**.

### Métricas resumidas
{tabla_md(df_resumen, ["Modelo", "Accuracy", "Precision", "Recall", "F1-score"])}

## 4️⃣ Variables más relevantes

### Regresión Logística
{tabla_md(coef_logreg_top, ["Variable", "Coeficiente"])}

### Random Forest
{tabla_md(importancia_rf_top, ["Variable", "Importancia"])}

## 5️⃣ Insights principales
- Tiempo en la empresa (`tenure`): Clientes nuevos o con poco tiempo tienen mayor riesgo de cancelar.
- Gasto mensual/total (`Charges.Monthly`, `Charges.Total`): Clientes con cargos elevados presentan más churn.
- Tipo de contrato y servicios contratados (`Contract`, `InternetService`, `OnlineSecurity`): Algunos planes o servicios aumentan la probabilidad de cancelación.

## 6️⃣ Estrategias de retención sugeridas
1. Incentivos para clientes nuevos: descuentos o promociones para reducir cancelación temprana.
2. Monitoreo de clientes con cargos altos: campañas personalizadas para explicar beneficios.
3. Optimización de servicios y contratos: mejorar experiencia de clientes en servicios críticos.
4. Alertas de riesgo de churn: usar el modelo predictivo para identificar clientes en riesgo y aplicar acciones proactivas.
"""

display(Markdown(informe_md))
```
